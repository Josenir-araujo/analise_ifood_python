```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
#Paremetros
mu = 5 # Mdia
sigma = 1 # Desvio padro
# Calcular a probabilidade de entrega em 6 dias ou menos
probabilidade = norm.cdf(6, loc=mu, scale=sigma)
#Visualizao da distribuio normal
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, loc=mu, scale=sigma)
plt.plot(x, y, label='Distribuio Normal')
plt.fill_between(x, y, where=(x <= 6), color='lightblue', alpha=0.3, label='rea at 6 dias')
plt.axvline(6, color='red', linestyle='--', label='6 dias de entrega')
plt.title('Distribuio Normal de Entrega')
plt.xlabel('Dias de Entrega')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid()
plt.show()
# Exibir a probabilidade calculada
print(f"A probabilidade de entrega em 6 dias ou menos : {probabilidade:.2%}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
# Parmetros
n = 50 # Nmero total de e-mails enviados 
p = 0.2 # Taxa de abertura de e-mails 
k = 15 # Nmero de e-mails que desejamos que sejam abertos
# Calcular a probabilidade de exatamente 15 e-mails serem abertos
probabilidade = binom.pmf(k, n, p)
# Visualizao da distribuio binomial
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)
plt.bar(x, y, color='lightblue', label='Distribuio Binomial')
plt.axvline(k, color='red', linestyle='--', label='15 e-mails abertos')
plt.title('Distribuio Binomial de Abertura de E-mails')
plt.xlabel('Nmero de E-mails Abertos')
plt.ylabel('Probabilidade')
plt.legend()
plt.grid()
plt.show()
# Exibir a probabilidade calculada
print(f"A probabilidade de exatamente 15 e-mails serem abertos : {probabilidade:.2%}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
# Parmetros
a = 0 # Limite inferior do intervalo
b = 10 # Limite superior do intervalo
# Calcular a probabilidade de um valor estar entre 2 e 8
probabilidade = uniform.cdf(8, loc=a, scale=b-a) - uniform.cdf(2, loc=a, scale=b-a)
# Visualizao da distribuio uniforme
x = np.linspace(a, b, 1000)
y = uniform.pdf(x, loc=a, scale=b-a)
plt.plot(x, y, label='Distribuio Uniforme')
plt.fill_between(x, y, where=((x >= 2) & (x <= 8)), color='lightblue', alpha=0.3, label='rea entre 2 e 8')
plt.axvline(2, color='red', linestyle='--', label='2')
plt.axvline(8, color='green', linestyle='--', label='8')
plt.title('Distribuio Uniforme de Valores')
plt.xlabel('Valores')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid()
plt.show()
# Exibir a probabilidade calculada
print(f"A probabilidade de um valor estar entre 2 e 8 : {probabilidade:.2%}")
```
